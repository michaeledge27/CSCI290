```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame( iris.data, columns = iris.feature_names)
```

```python
df
```

```python
df["species"] = iris.target
```

```python
target_names = {      # switching to 0, 1, 2 for convenience
    0:'setosa',
    1:'versicolor',
    2:'virginica'
}

df['species_name'] = df['species'].map(target_names)
df.head()
```

```python
plt.figure(figsize=(5, 3))
sns.scatterplot( x = "petal length (cm)", y = "petal width (cm)", data = df, hue = "species_name");
```

```python
# find the averages of the petal lengths (x's) and the petal widths (y's) for each of the species
centroids = df.groupby("species_name")[["petal length (cm)", "petal width (cm)"]].mean()
```

```python
plt.figure(figsize=(5, 3))
sns.scatterplot( x = "petal length (cm)", y = "petal width (cm)", data = df, hue = "species_name")
sns.scatterplot( x = "petal length (cm)", y = "petal width (cm)", data = centroids, c = "black", marker="X", s = 100);
```
